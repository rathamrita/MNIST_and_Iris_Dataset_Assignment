```python
df.loc[df["Species"]=="Iris-setosa","Species"]=0
df.loc[df["Species"]=="Iris-versicolor","Species"]=1
df.loc[df["Species"]=="Iris-virginica","Species"]=2
print(df.head())
```

```python
df=df.iloc[np.random.permutation(len(df))]
print(df.head())
```

```python
X=df.iloc[:,1:5].values
y=df.iloc[:,5].values

print("Shape of X",X.shape)
print("Shape of y",y.shape)
print("Examples of X\n",X[:3])
print("Examples of y\n",y[:3])
```

```python
X_normalized=normalize(X,axis=0)
print("Examples of X_normalised\n",X_normalized[:3])
```

### 4. Data splitting into train and test set

```python
total_length=len(df)
train_length=int(0.8*total_length)
test_length=int(0.2*total_length)

X_train=X_normalized[:train_length]
X_test=X_normalized[train_length:]
y_train=y[:train_length]
y_test=y[train_length:]

print("Length of train set x:",X_train.shape[0],"y:",y_train.shape[0])
print("Length of test set x:",X_test.shape[0],"y:",y_test.shape[0])
```

### 5. Model Building

```python
from keras.utils import np_utils
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, BatchNormalization
from sklearn.metrics import f1_score
```

```python
y_train=np_utils.to_categorical(y_train,num_classes=3)
y_test=np_utils.to_categorical(y_test,num_classes=3)
print("Shape of y_train",y_train.shape)
print("Shape of y_test",y_test.shape)
```

```python
from tensorflow.keras.layers import Dense, Dropout
model=Sequential()
model.add(Dense(1000,input_dim=4,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(3,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size=20,epochs=15,verbose=1)
```

### 6.Model Evaluation

```python
prediction=model.predict(X_test)
length=len(prediction)
y_label=np.argmax(y_test,axis=1)
predict_label=np.argmax(prediction,axis=1)

accuracy=np.sum(y_label==predict_label)/length * 100 
print("Accuracy of the dataset",accuracy )
```

```python
import matplotlib.pyplot as plt

history = model.fit(X_train, y_train, epochs=15, batch_size=20, validation_data=(X_test, y_test), verbose=1)
# Get the accuracy values from the training history
accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']

# Get the number of epochs
epochs = range(1, len(accuracy) + 1)

# Plot the accuracy over epochs
plt.plot(epochs, accuracy, 'b', label='Training Accuracy')
plt.plot(epochs, val_accuracy, 'r', label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

### 7. Model Interpretation
